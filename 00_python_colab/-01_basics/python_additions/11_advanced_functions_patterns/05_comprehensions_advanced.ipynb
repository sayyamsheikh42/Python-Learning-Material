## Exercises
1) Flatten nested list  2) Word frequency dict

```python
matrix=[[1,2,3],[4,5,6]]
flat=[x for row in matrix for x in row]
print(flat)

text='to be or not to be'
freq={w: sum(1 for t in text.split() if t==w) for w in set(text.split())}
print(freq)
```
