# Class Decorators — Theory & Practice
A class decorator takes a class and returns a (possibly modified) class.

**Use cases**: registries, auto methods, invariants.

## Exercises
1) @auto_repr  2) @register

```python
def auto_repr(cls):
    def __repr__(self):
        kv = ', '.join(f"{k}={v!r}" for k,v in self.__dict__.items())
        return f"{cls.__name__}({kv})"
    cls.__repr__ = __repr__
    return cls

@auto_repr
class Person:
    def __init__(self, name, age):
        self.name, self.age = name, age

print(Person('Aisha', 22))
```

```python
REGISTRY = {}

def register(cls):
    REGISTRY[cls.__name__] = cls
    return cls

@register
class Foo: pass

@register
class Bar: pass

print(REGISTRY)
```
