# Closures — Theory & Practice
A closure is a function that remembers values from its lexical scope even after the outer function returns.

**Why useful**: stateful functions, configuration, encapsulation.

## Exercises
1) Counter with step & reset
2) Length validator
3) Timer closure

```python
def make_counter(start=0):
    value = start
    def inc(step=1):
        nonlocal value
        value += step
        return value
    return inc

c = make_counter(10)
print(c(), c(), c(5))
```

```python
def make_length_validator(min_len=0, max_len=999):
    def validate(s: str) -> bool:
        return min_len <= len(s) <= max_len
    return validate

v = make_length_validator(2, 4)
print(v('hi'), v('hello'))
```
