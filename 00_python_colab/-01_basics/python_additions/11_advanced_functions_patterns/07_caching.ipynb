# Caching — functools.lru_cache
## Exercises
1) Compare fib with/without cache  2) Cache slow I/O

```python
import time
from functools import lru_cache

def fib_plain(n):
    return n if n < 2 else fib_plain(n-1) + fib_plain(n-2)

@lru_cache(maxsize=None)
def fib_fast(n):
    return n if n < 2 else fib_fast(n-1) + fib_fast(n-2)

start=time.time(); print(fib_plain(25)); print('plain:', time.time()-start)
start=time.time(); print(fib_fast(35));  print('cached:', time.time()-start)
```
