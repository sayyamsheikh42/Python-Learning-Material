# Project 01 — CLI Log Analyzer
Skills: argparse, file I/O, error handling.

```python
import argparse, re
from collections import Counter

def analyze(path, level=None, top=5):
    counts = Counter(); total = 0
    level_pat = re.compile(r'\b(INFO|WARN|ERROR)\b')
    with open(path, encoding='utf-8', errors='ignore') as f:
        for line in f:
            total += 1
            m = level_pat.search(line)
            if m:
                lvl = m.group(1)
                if level and lvl != level:
                    continue
                counts[lvl] += 1
    return total, counts.most_common(top), counts

# Example usage (as a script):
# python log_analyzer.py --path sample.log --level ERROR --top 3
```
