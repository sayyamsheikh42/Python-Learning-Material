# Project 03 — Data Validation Pipeline (Dataclasses vs Pydantic)

```python
from dataclasses import dataclass
from typing import Optional

@dataclass(frozen=True)
class UserDC:
    id: int
    name: str
    email: Optional[str] = None

try:
    from pydantic import BaseModel, EmailStr
    class UserModel(BaseModel):
        id: int
        name: str
        email: Optional[EmailStr] = None
    HAVE_PYDANTIC = True
except Exception:
    HAVE_PYDANTIC = False

record = {'id': 1, 'name': 'Aisha', 'email': 'aisha@example.com'}
user_dc = UserDC(**record)
print(user_dc)

if HAVE_PYDANTIC:
    print(UserModel(**record))
else:
    print('Pydantic not installed; skipping pydantic example.')
```
