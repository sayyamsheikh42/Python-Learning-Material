# Project 05 — Algorithm + Caching Demo
Compare recursive, iterative, and cached Fibonacci.

```python
import time
from functools import lru_cache

def fib_rec(n):
    return n if n<2 else fib_rec(n-1)+fib_rec(n-2)

def fib_iter(n):
    a,b=0,1
    for _ in range(n):
        a,b=b,a+b
    return a

@lru_cache(maxsize=None)
def fib_cached(n):
    return n if n<2 else fib_cached(n-1)+fib_cached(n-2)

def bench(fn, n=35, reps=1):
    start=time.time(); out=None
    for _ in range(reps):
        out=fn(n)
    return out, time.time()-start

for fn in (fib_rec, fib_iter, fib_cached):
    out, t = bench(fn, n=35)
    print(fn.__name__, '=>', out, 'time=', round(t,3), 's')
```
