# Project 02 — Async API Fetcher (Mocked)
Skills: asyncio, tasks, timeouts, error handling.

```python
import asyncio, random, time

async def mock_fetch(name, delay=None):
    delay = delay or random.uniform(0.2, 1.2)
    await asyncio.sleep(delay)
    return {'name': name, 'delay': round(delay,2)}

async def main():
    tasks = [mock_fetch(f'ep{i}') for i in range(6)]
    start = time.time()
    results = await asyncio.gather(*tasks)
    print('results:', results, 'elapsed:', round(time.time()-start,2), 's')

# asyncio.run(main())
```
